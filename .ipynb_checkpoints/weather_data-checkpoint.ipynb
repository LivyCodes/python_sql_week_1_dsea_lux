```python
#importing pandas in my current db that i'm working with
import pandas as pd
```

```python
#importing numpy in my current db that i'm working with
import numpy as np
```

```python
#reading the csv into my database
weather_data_1=pd.read_csv("c:/users/jonar/data/weather_data.csv")
```

```python
#reading first 10 records of the weather dataset
weather_data_1.head(10)
```

```python
#finding all records where the weather was exactly clear
clear_weather = weather_data_1[weather_data_1['Weather']=='Clear']

clear_weather
```

```python
#Finding the number of times the wind speed was exactly 4 km/hr
wind_speed = weather_data_1[weather_data_1['Wind Speed_km/h'] == 4]

wind_speed['Wind Speed_km/h'].count()
```

```python
#checking if there are any null values present in the dataset
weather_data_1.info()
null_info = weather_data_1.isnull().sum()
print(null_info)
```

```python
#renaming 'Weather' to 'Weather_Condition'
weather_data_1.rename(columns={"Weather":"Weather_Condition"})
```

```python
#mean visibility of the dataset
weather_data_1['Visibility_km'].mean()
```

```python
#number of records where windspeed is greater than 24 km/hr and visibility equal to 25km
windspeed_visibility_check = weather_data_1[(weather_data_1['Wind Speed_km/h'] > 24) & (weather_data_1['Visibility_km'] == 25)]
windspeed_visibility_check.count()
```

```python
#mean value of each column for each weather condition
mean_temp = weather_data_1['Temp_C'].mean()
mean_Dew_Point_Temp_C =weather_data_1['Dew Point Temp_C'].mean()
mean_Rel_Hum = weather_data_1['Rel Hum_%'].mean()
mean_Wind_Speed = weather_data_1['Wind Speed_km/h'].mean()
mean_Visibility_km = weather_data_1['Visibility_km'].mean()
mean_Press_kPa = weather_data_1['Press_kPa'].mean()
print('Mean_Temp_C: ', mean_temp)
print('Mean_Dew_Point_Temp_C: ', mean_Dew_Point_Temp_C )
print('Mean_Rel_Hum_%: ', mean_Rel_Hum )
print('Mean_Wind_Speed: ', mean_Wind_Speed)
print('Mean_Visibility_km: ', mean_Visibility_km )
print('Mean_Press_kPa: ', mean_Press_kPa)
```

```python
#instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40
clear_humidity_visibility=weather_data_1[(weather_data_1['Weather']== 'Clear') & ((weather_data_1['Rel Hum_%'] > 50) | (weather_data_1['Visibility_km'] > 40))]
clear_humidity_visibility
```

```python
#Find the number of weather conditions that include snow
snowy_weather = weather_data_1 [weather_data_1['Weather'] == 'Snow']
snowy_weather['Weather'].count()
```
